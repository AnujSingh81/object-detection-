```python
from ultralytics import YOLO
```

```python
model =YOLO("yolo11m-seg.pt")
```

```python
results = model("/content/bus.jpg")
```

```python
for result in results:
  xy= result.masks.xy # make polygon
  xyn= result.masks.xyn # do normalization
  mask = result.masks.data #create a matric
  print(xy)
  print(xyn)
  print(mask)
```

```python
mask_img = result.plot()
```

```python
result.save("mask_img.jpg")
```
